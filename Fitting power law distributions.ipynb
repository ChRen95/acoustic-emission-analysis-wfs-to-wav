Power law distribution
====================

Taken from [http://dx.doi.org/10.1137/070710111].

$$P(x\leq X<x+\text{d}x) = C x^{-\alpha} $$

$$ 1 = \int _ {x_\text{min}} ^\infty C x^{-a} \text{d}x 
= C \left[ \frac{x^{-\alpha+1}}{-\alpha+1} \right]_{x_\text{min}}^\infty
= C \left( 0 - \frac{x_\text{min}^{-\alpha+1}}{-\alpha+1} \right); \qquad
C = \frac{\alpha-1}{x_\text{min}^{-\alpha+1}} $$

$$P(x\leq X<x+\text{d}x) = \frac{\alpha-1}{x_\text{min}} \left( \frac{x}{x_\text{min}}\right)^{-\alpha} $$

Cumulative distribution function
--------------------------------

$$CDF(x) = P(x\leq X) = \int_{x'}^\infty C {x'}^{-\alpha} \text{d}x' = \left( \frac{x}{x_\text{min}}\right)^{-\alpha+1}$$

Random number generation
------------------------

$$X_\text{power law} = CDF^{-1}\left(1-X_\text{uniform}\right) = x_\text{min} \left(1-X_\text{uniform}\right)^{\frac{1}{-\alpha+1}}$$

```python
def random_power(xmin,a, size=1):
    return xmin*(1.-random.uniform(size=size))**(1./(a+1))
```


Maximum likelihood estimator of $\alpha$
------------------------------------

$$\mathcal{L} = \ln P(x\mid\alpha) 
= \ln \prod_i \frac{\alpha-1}{x_\text{min}} \left( \frac{x_i}{x_\text{min}}\right)^{-\alpha} $$ 
$$ = \sum_i \left[ \ln(\alpha-1) - \ln(x_\text{min}) -\alpha \ln\frac{x_i}{x_\text{min}} \right] $$
$$ = n \ln(\alpha-1) - n \ln(x_\text{min}) -\alpha \sum_i \ln\frac{x_i}{x_\text{min}} $$

$$ 0 = \frac{\partial \mathcal{L}}{\partial\alpha} = \frac{n}{\alpha-1} - \sum_i \ln\frac{x_i}{x_\text{min}} $$
$$ \hat{\alpha} = 1+n\left[ \sum_i \ln\frac{x_i}{x_\text{min}} \right]^{-1}$$


```python
def mle(xmin, data):
    d = data[data>=xmin]
    a = 1 + data.size/sum(log(d/xmin))
    return -a, (a-1)/sqrt(d.size)    
```

Exponential bins
---------------

Make the histogram with bin sizes exponentially increasing so that they appear
equal in log-log plot.

```python
def loghist(data, bins=50, doplot=False):
    bins = exp(linspace(log(data.min()), log(data.max()), bins))
    a,b = histogram(data, bins=bins, density=True)

    y = log( a[a>0] )
    x = log( (b[1:]+b[:-1])/2 )[a>0]

    p = poly1d(polyfit(x,y,1))
    
    if doplot:
        plot(x, y, "o")
        plot(x,p(x), label=p.coeffs[0])
        
    return p.coeffs[0]
    

def histlog(data, bins=50, doplot=False):
    a,b = histogram(log(data), bins=bins, density=False)

    y = log( (a/diff(exp(b))/data.size)[a>0] )
    x = ( (b[1:]+b[:-1])/2 )[a>0]

    p = poly1d(polyfit(x,y,1))
    
    if doplot:
        plot(x, y, "o")
        plot(x,p(x), label=p.coeffs[0])
        
    return p.coeffs[0]
```

Joined bins
----------

Make classic histogram, and join neighbouring bins until they reach some minimum amount of counts.

```python
def joinbins(a,b, mincount=10):
    newb, newa = [b[0]], []
    s = 0
    for aa,bb in zip(a,b[1:]):
        s += aa
        if s < mincount:
            continue
        newa.append(s)
        newb.append(bb)
        s = 0
    if s > 0:
        newa.append(s)
        newb.append(bb)
    return asarray(newa), asarray(newb)

def joinhist(data, bins=500, mincount=10, doplot=False):
    a,b = histogram(data, bins=bins, density=False)
    a,b = joinbins(a,b,mincount)
    
    y = log( a/diff(b)/data.size )[a>0]
    x = log( (b[1:]+b[:-1])/2 )[a>0]

    p = poly1d(polyfit(x,y,1))
    a = p.coeffs[0]
    
    if doplot:
        plot(x, y, "o")
        plot(x,p(x),label=p.coeffs[0])
        
    return p.coeffs[0]
```

Cumulative distribution function
-----------------------------
http://en.wikipedia.org/wiki/Power_law#Plotting_power-law_distributions

Make log(data) vs. log(rank) plot. 
CDF is again power-law function with exponent increased by one.

```python
def cdf(data, doplot=False):
    x = log( sort(data) )
    y = log( arange(data.size,0,-1) )
    p = poly1d(polyfit(x,y,1))
    
    if doplot:
        plot(x, y, "o", ms=3)
        plot(x,p(x),label=p.coeffs[0])
        
    return p.coeffs[0]-1
```

Test it
======

```python
xmin, a = 1, -2.5
data = random_power(xmin, a, 50000)

figure(figsize=(12,8))

subplot(221)
print loghist(data, bins=50,  doplot=True)
legend()

subplot(222)
print histlog(data, bins=50,  doplot=True)
legend()

subplot(223)
print joinhist(data, bins=500, mincount=10,  doplot=True)
legend()

subplot(224)
print cdf(data, doplot=True)
legend()

print mle(xmin, data)
```

```python
funcs = [loghist, histlog, joinhist, cdf, lambda data: mle(xmin, data)[0]]
funcs[-1].__name__ = "mle"
sizes = range(100, 15000,750)

for f in funcs:
    v,e = [], []   
    for s in sizes:
        d = []
        for i in range(50):
            data = random_power(xmin, a, s)
            d.append(f(data))
        v.append(mean(d))
        e.append(std(d))
    errorbar(sizes, v, e, label=f.__name__)
axhline(a,c="k",ls="--")
legend();        
```

```python
params = linspace(-3,-1.1,11)

for f in funcs:
    v,e = [], []
    for a in params:
        d = []
        for i in range(50):
            data = random_power(xmin, a, 10000)
            d.append(f(data))
        v.append(mean(d))
        e.append(std(d))
    errorbar(params, v, e, label=f.__name__)
plot(params,params,c="k",ls="--")
autoscale(tight=True)
legend(loc=2);               
```
